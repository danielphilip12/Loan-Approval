## Feature Engineering and Reorganization

```python
df_app.head()
```

```python
df_data.head()
```

We will change df_app's columns `Marital_Status` and `Credit_Score` to match the format of df_data, that is, change Single  to No (not married), and Married to yes (married). We will also convert `Credit_Score` to be 1s and 0s, with a score of 670 and above 1, 669 and below 0. 

`Loan_Approved` columns are also different, in that one says Yes/No, and the other is Y/N. We will change this as well. 

```python
df_app['Loan_Term'].describe()
```

```python
df_data['Loan_Term'].describe()
```

Also, looking at `Loan_Term`, the units may not be the same in both. So we should drop it to be cautious. 

```python
df_app = df_app.drop(columns=['Loan_Term'])
df_data = df_data.drop(columns=['Loan_Term'])
```

```python
df_app['Marital_Status'] = df_app['Marital_Status'].map({'Married': 'Yes', 'Single': 'No'})
df_app['Loan_Approved'] = df_app['Loan_Approved'].map({'Yes': 'Y', 'No': 'N'})
```

```python
def good_credit_score(score):
    if score >= 670:
        return 1.0
    return 0.0
```

```python
df_app['Credit_History'] = df_app['Credit_Score'].apply(good_credit_score)
```

```python
df_app = df_app.drop(columns=['Credit_Score'])
```

```python
df_app.head()
```

```python
column_order = df_app.columns
```

```python
df_data = df_data[column_order]
```
